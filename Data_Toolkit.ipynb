Q.2. How does broadcasting work in NumPy?

Ans:In NumPy, broadcasting is a mechanism that allows array operations to be performed on arrays of different shapes in an element-wise fashion without explicitly copying data. It enables efficient computation by automatically expanding smaller arrays to match the shape of larger ones when performing arithmetic operations.

Broadcasting Rules:

For two arrays to be broadcasted together, NumPy follows these rules:

1. Compare shapes from right to left

  -The dimensions must be either the same or one of them must be 1.

  -If a dimension is 1, it is stretched to match the corresponding dimension of the other array.
2. If the dimensions do not match and neither is 1, broadcasting is not possible, and NumPy raises a ValueError.

**Example of broadcasting:**




```python
import numpy as np

A = np.array([[1, 2, 3], [4, 5, 6]])  # Shape: (2, 3)
B = np.array([1, 2, 3])               # Shape: (3,)

C = A + B  # Broadcasting occurs
print(C)
```

Q.17.  Explain the difference between apply() and map() in Pandas.

Ans: Both apply() and map() are used to apply functions to Pandas DataFrames and Series, but they have key differences in how they work.

**map() – Works Only on Series (Single Column)**

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({"Name": ["Alice", "Bob", "Charlie"], "Salary": [50000, 60000, 70000]})
print(df)
# Apply function using map()
df["Salary"] = df["Salary"].map(lambda x: x * 1.10)  # Increase salary by 10%

print(df)
```

**apply() – Works on Both Series & DataFrames**

Example 1: Using apply() on a Series

```python
df["Salary"] = df["Salary"].apply(lambda x: x * 1.10)
print(df)
```

Using apply() on a DataFrame


```python
# Apply function across all rows
df["Salary After Tax"] = df.apply(lambda row: row["Salary"] * 0.85 if row["Salary"] > 60000 else row["Salary"], axis=1)

print(df)
```

Q.19. How does Pandas simplify time series analysis

Ans:Pandas provides powerful tools to work with time series data, making it easy to handle date-based indexing, resampling, rolling statistics, and time zone conversions.

 1. DateTime Indexing & Conversion:
Pandas can automatically parse dates and use them as an index, making time-based operations easy.

##Example: Creating a DateTime Index

```python
import pandas as pd

# Creating a time series
dates = pd.date_range(start="2024-01-01", periods=5, freq="D")
data = [10, 20, 15, 30, 25]
df = pd.DataFrame({"Value": data}, index=dates)

print(df)
```

2. Resampling (Changing Time Granularity):
Resampling is used to aggregate or downsample/up-sample time series data.

## Example: Resampling Data to Weekly Frequency

```python
df_resampled = df.resample("W").mean()  # Convert daily data to weekly averages
print(df_resampled)
```

3. Shifting & Lagging (Time-Based Operations):
Pandas can shift time series data forward or backward to analyze trends.

##Example: Shifting Data

```python
df["Previous Day"] = df["Value"].shift(1)  # Shift values by 1 day
print(df)
```

4. Rolling Windows (Moving Averages):
Rolling statistics allow smoothing time series data by computing moving averages.
##Example: 3-Day Moving Average

```python
df["3-Day MA"] = df["Value"].rolling(window=3).mean()
print(df)
```

5. Handling Missing Dates & Data:
Pandas provides tools to fill or interpolate missing values in time series.
##Example: Filling Missing Dates

```python
df_missing = df.asfreq("D")  # Ensure daily frequency
df_filled = df_missing.fillna(method="ffill")  # Forward fill missing values
print(df_filled)
```

Q.21. Why is NumPy’s array slicing faster than Python’s list slicing?

Ans:NumPy’s array slicing is significantly faster and more memory-efficient compared to Python’s list slicing because of its underlying implementation and data structure.

 1. NumPy Uses Continuous Memory Allocation:
NumPy stores elements in a contiguous block of memory (C-style or Fortran-style).
Python lists store references to objects scattered across memory.

2. NumPy Uses Views, Not Copies:
NumPy slicing creates a "view" (a reference) of the original array, not a new copy.

 Python lists create a new list when sliced, increasing memory usage and time complexity.

3. NumPy is Optimized with C and Vectorized Operations:
NumPy operations run in C, while Python lists require high-level Python loops, which are slow.

 NumPy takes advantage of SIMD (Single Instruction, Multiple Data) and parallel computing.

4. NumPy Uses Fixed Data Types (Less Overhead):
NumPy arrays store data as a single type (e.g., int32, float64), leading to less memory overhead.

 Python lists store heterogeneous data (each element is an object reference).
